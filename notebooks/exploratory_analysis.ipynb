```python
import pandas as pd
import numpy as np   
import os
import matplotlib.pyplot as plt

import seaborn as sns
```

I. Statistiques descriptives

```python
print(os.getcwd())

df = pd.read_csv("../data/raw/ecommerce_messy_data.csv")
df.head()
df.describe()
df.info()
df.head()


```

```python
df.hist(figsize=(12,8), bins=20, color='lightblue', edgecolor='black')
plt.tight_layout()
plt.show()


# Heatmap pour voir la corr

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélations")
plt.show()


# Visualiser deux var numériques

sns.boxplot(x="Quantity", y="TotalAmount", data=df)
plt.xticks(rotation=45)
plt.show()


sns.boxplot(x="ProductCategory", y="TotalAmount", data=df)
plt.xticks(rotation=45)
plt.show()


# Distribution des var catégorielles

df["CustomerSegment"].value_counts().plot(kind="bar", color="orange")
plt.show()

df["ProductCategory"].value_counts().plot(kind="bar", color="blue")
plt.show()
```
